**1.Cargue en un Dataframe los datos de los accidentes de los últimos 3 años**

```python
df2020 = pd.read_csv("/kaggle/input/datasetdechoques/input/atus_anual_2020.csv", index_col=False)
df2021 = pd.read_csv("/kaggle/input/datasetdechoques/input/atus_anual_2021.csv", index_col=False)
df2022 = pd.read_csv("/kaggle/input/datasetdechoques/input/atus_anual_2022.csv", index_col=False)
dfAccidentes = pd.concat([df2020, df2021, df2022])
dfAccidentes
```

Se notó que los índices se estaban repitiendo, por lo tanto, es necesario abordar ese problema y resolverlo.

```python
dfAccidentes = dfAccidentes.reset_index(drop=True)
dfAccidentes
```

**2.Agregue una columna tipo date que englobe las columnas correspondientes a Año, Mes, Dia, Hora y Minuto.**



```python
dfAccidentes['FECHA'] = pd.to_datetime({'year': dfAccidentes["ANIO"],
                                       'month': dfAccidentes["MES"],
                                       'day': dfAccidentes["ID_DIA"],
                                       'hours': dfAccidentes["ID_HORA"],
                                       'minutes': dfAccidentes["ID_MINUTO"]})
dfAccidentes['FECHA']
```

**3.Agregue una columna para Área cuyos valores serán Urbana o Suburbana dependiendo del área donde ocurrió el accidente.**

es mas sencillo crear una funcion y luego aplicarla

```python
#funcion para definir si el area fue urbana o suburbana
def fill_area(dato):
    if 'Sin accidente en esta zona' not in dato["URBANA"]:
        return 'Urbana'
    elif 'Sin accidente en esta zona' not in dato["SUBURBANA"]:
        return 'Suburbana'
    else:
        return 'NA'
print("Hecho")
```

se le apica la funcion a la nueva columna para llenar algunos datos

```python
#axis=1 para especificar columnas
dfAccidentes['AREA'] = dfAccidentes.apply(lambda dato: fill_area(dato), axis=1)
dfAccidentes['AREA']
```

**4.Agregue una columna Zona cuyos valores serán la zona donde ocurrió el accidente**

para agregar una nueva columna de ZONA podemos utilizar una funcion similar a la anterior con la unica diferencia de que retorne los valores de dichas columnas y no una cadena predeterminada

```python
#funcion para definir la zona
def fill_zona(dato):
    if 'Sin accidente en esta zona' not in dato["URBANA"]:
        return dato["URBANA"]
    elif 'Sin accidente en esta zona' not in dato["SUBURBANA"]:
        return dato["SUBURBANA"]
    else:
        return 'NA'
print("Hecho")
```

funcion de la nueva colum

```python
dfAccidentes['ZONA'] = dfAccidentes.apply(lambda dato: fill_zona(dato), axis=1)
dfAccidentes['ZONA']
```

**5.Elimine las columnas URBANA y SUBURBANA**



```python
dfAccidentes = dfAccidentes.drop(['URBANA', 'SUBURBANA'], axis=1)
dfAccidentes
```

**9.¿Cómo se compara el número de accidentes de cada año contra el anterior? ¿En qué porcentaje varió?**

```python
año_2020 = len(df2020)
año_2021 = len(df2021)
año_2022 = len(df2022)

accidentes_año = dfAccidentes["ANIO"].value_counts().reset_index(name='ACCIDENTES')
accidentes_año
```

Sacar la variacion:

```python
def variacion(dato):
    if dato.name < len(accidentes_año) - 1:
        año_act = dato["ACCIDENTES"]
        año_ant = accidentes_año.at[dato.name + 1, "ACCIDENTES"]
        variacion = ((año_act-año_ant)/año_ant)*100
        return variacion
    else: 
        return None

print("Finalizo")
```

**como podemos ver el porcentaje de variacion de un año con respecto al anterior es positivo y alrededor del 10 porciento**


```python
accidentes_año['VARIACION'] = accidentes_año.apply(variacion, axis=1)
accidentes_año
```
